```python
def factorial(n):
    fact = 1
    for i in range(2,n+1):
        fact*=i
    return fact
    def ncr(n,r):
        return(factorial(n)//factorial(r)*factorial(n-r))
    def binomial_coefficients(n):
        return([ncr(n,r)for r in range(n+1)])
        n=10
        print(binomial_coefficients(n))
```
